```python
__author__ = 'joncomo'


class Pair(object):
    already_created_pairs = set()

    def __init__(self, i, j):
        self.i = i
        self.j = j

    def out(self):
        return 2**self.i * 5**self.j

    @staticmethod
    def get_unique(i, j):
        if Pair.already_created_pairs.isdisjoint({(i, j)}):
            Pair.already_created_pairs.add((i, j))
            return Pair(i, j)

    def children(self):
        returnChildren = []
        left = Pair.get_unique(self.i + 1, self.j)
        right = Pair.get_unique(self.i, self.j + 1)
        if left:
            returnChildren.append(left)
        if right:
            returnChildren.append(right)
        return returnChildren

pairs = [Pair(0,0)]

#arbitrary n count output
n = 0
while n < 100:
    smallestPair = pairs[0]
    smallestOut = smallestPair.out()

    for p in pairs:
        if p.out() < smallestOut:
            smallestOut = p.out()
            smallestPair = p

    print "next out (%i,\t%i)\t= %i" % (smallestPair.i, smallestPair.j, smallestPair.out())
    pairs.remove(smallestPair)
    pairs.extend(smallestPair.children())
    n += 1
```
